```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) #20 percent #42 same split 
```

```python
x_train
```

```python
#linear eq
from sklearn.linear_model import LinearRegression
```

```python
#variable lr_model
lr_model=LinearRegression()
```

```python
#Train the model
lr_model.fit(x_train,y_train) #it fit model
```

```python
#predict the model
predict=lr_model.predict(x_test)
```

```python
predict
```

```python
result=pd.DataFrame({'Actual_value':y_test,'predicted_by_model':predict})
result
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
r2_score=r2_score(y_test,predict)
r2_score
#how well a model (like a math equation) predicts the correct answers.
```

```python
#MSE tells us if the model sometimes makes wildly bad predictions (even if most are decent).
```

```python
mse=mean_squared_error(y_test,predict)
mse
```

```python
# MAE tells us the typical error size (easier to explain to non-experts).
```

```python
mae=mean_absolute_error(y_test,predict)
mae
```
